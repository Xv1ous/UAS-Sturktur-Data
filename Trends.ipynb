```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

Graph dari data Instagram

```python
# Pembacaan CSV file DataFrame Instagram
df_instagram = pd.read_csv("instagram_trends.csv")

# Convert the 'date' column to datetime type
df_instagram["date"] = pd.to_datetime(df_instagram["date"])

# Penampilan data frame Instagram
print(df_instagram.head())

# Plot Instagram data
plt.figure(figsize=(10, 6))
plt.plot(df_instagram["date"], df_instagram["Instagram"], label="Instagram", color="orange")
plt.xlabel("Date")
plt.ylabel("Instagram")
plt.title("Instagram Data")
plt.legend()
plt.grid()
plt.show()
```

Skor Instagram Seiring Waktu: Kumpulan data menunjukkan skor Instagram dari 17 Juli 2022 hingga 9 Juli 2023. Skor tampaknya berfluktuasi selama periode ini.

     Skor Tinggi: Ada beberapa kejadian di mana skor Instagram mencapai atau melebihi 95, yaitu pada 17/7/2022, 24/7/2022, 14/8/2022, 2/10/2022, 23/10/2022, 10 /30/2022, 13/11/2022, 18/12/2022, 25/12/2022, 22/1/2023, 19/3/2023, dan 2/7/2023.

     Skor Rendah: Skor terendah tampaknya sekitar 77, diamati pada 20/11/2022, 27/11/2022, dan 4/12/2022.

     Fluktuasi: Skor umumnya berfluktuasi antara poin tinggi dan rendah, yang menunjukkan variasi keterlibatan atau popularitas dari waktu ke waktu.

     Data Parsial: Ada kolom "isPartial" yang menunjukkan apakah data sebagian atau tidak. Entri terakhir pada 7/9/2023 ditandai sebagai "BENAR", menunjukkan bahwa data mungkin tidak lengkap untuk tanggal tersebut.

## Korelasi

```python


# Calculate correlation for TikTok DataFrame
correlation_tiktok = df_tiktok.corr(numeric_only=True)
print("Correlation Matrix for TikTok:")
print(correlation_tiktok)

# Calculate correlation for YouTube DataFrame
correlation_youtube = df_youtube.corr(numeric_only=True)
print("\nCorrelation Matrix for YouTube:")
print(correlation_youtube)

# Calculate correlation for Twitter DataFrame
correlation_twitter = df_Twitter.corr(numeric_only=True)
print("\nCorrelation Matrix for Twitter:")
print(correlation_twitter)

# Calculate correlation for Facebook DataFrame
correlation_facebook = df_Facebook.corr(numeric_only=True)
print("\nCorrelation Matrix for Facebook:")
print(correlation_facebook)

# Calculate correlation for Instagram DataFrame
correlation_instagram = df_instagram.corr(numeric_only=True)
print("\nCorrelation Matrix for Instagram:")
print(correlation_instagram)
```

```python
# Set up the subplots
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# Plot the correlation matrix for TikTok
sns.heatmap(correlation_tiktok, annot=True, cmap="coolwarm", ax=axes[0, 0])
axes[0, 0].set_title("Correlation Matrix for TikTok")

# Plot the correlation matrix for YouTube
sns.heatmap(correlation_youtube, annot=True, cmap="coolwarm", ax=axes[0, 1])
axes[0, 1].set_title("Correlation Matrix for YouTube")

# Plot the correlation matrix for Twitter
sns.heatmap(correlation_twitter, annot=True, cmap="coolwarm", ax=axes[1, 0])
axes[1, 0].set_title("Correlation Matrix for Twitter")

# Plot the correlation matrix for Facebook
sns.heatmap(correlation_facebook, annot=True, cmap="coolwarm", ax=axes[1, 1])
axes[1, 1].set_title("Correlation Matrix for Facebook")

# Adjust layout
plt.tight_layout()

# Show the combined graph
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you have already loaded the data into DataFrames as you mentioned
# df_tiktok, df_youtube, df_Twitter, df_Facebook, df_instagram

# Function to plot line graph
def plot_line_graph(df, platform_name):
    plt.plot(df['date'], df[platform_name], label=platform_name)

# Set up the figure and plot each line graph
plt.figure(figsize=(10, 6))

# Plot Instagram data
plot_line_graph(df_instagram, "Instagram")

# Plot Facebook data
plot_line_graph(df_Facebook, "Facebook")

# Plot Twitter data
plot_line_graph(df_Twitter, "Twitter")

# Plot YouTube data
plot_line_graph(df_youtube, "Youtube")

# Plot TikTok data
plot_line_graph(df_tiktok, "TikTok")

# Set axis labels, title, and legend
plt.xlabel("Date")
plt.ylabel("Trend")
plt.title("Trends from Different Platforms")
plt.legend()

# Display the combined line graph
plt.grid()
plt.show()
```

```python
def plot_scatter_graph(df, platform_name):
    plt.scatter(df['date'], df[platform_name], label=platform_name)

# Set up the figure and plot each scatter graph
plt.figure(figsize=(10, 6))

# Plot Instagram data
plot_scatter_graph(df_instagram, "Instagram")

# Plot Facebook data
plot_scatter_graph(df_Facebook, "Facebook")

# Plot Twitter data
plot_scatter_graph(df_Twitter, "Twitter")

# Plot YouTube data
plot_scatter_graph(df_youtube, "Youtube")

# Plot TikTok data
plot_scatter_graph(df_tiktok, "TikTok")

# Set axis labels, title, and legend
plt.xlabel("Date")
plt.ylabel("Trend")
plt.title("Trends from Different Platforms (Scatter Plot)")
plt.legend()

# Display the combined scatter graph
plt.grid()
plt.show()
```
