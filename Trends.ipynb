```python
import pandas as pd
import matplotlib.pyplot as plt
```

Graph dari data Instagram

```python
# Pembacaan CSV file DataFrame Instagram
df_instagram = pd.read_csv("instagram_trends.csv")

# Convert the 'date' column to datetime type
df_instagram["date"] = pd.to_datetime(df_instagram["date"])

# Penampilan data frame Instagram
print(df_instagram.head())

# Plot Instagram data
plt.figure(figsize=(10, 6))
plt.plot(df_instagram["date"], df_instagram["Instagram"], label="Instagram", color="orange")
plt.xlabel("Date")
plt.ylabel("Instagram")
plt.title("Instagram Data")
plt.legend()
plt.grid()
plt.show()
```

Skor Instagram Seiring Waktu: Kumpulan data menunjukkan skor Instagram dari 17 Juli 2022 hingga 9 Juli 2023. Skor tampaknya berfluktuasi selama periode ini.

     Skor Tinggi: Ada beberapa kejadian di mana skor Instagram mencapai atau melebihi 95, yaitu pada 17/7/2022, 24/7/2022, 14/8/2022, 2/10/2022, 23/10/2022, 10 /30/2022, 13/11/2022, 18/12/2022, 25/12/2022, 22/1/2023, 19/3/2023, dan 2/7/2023.

     Skor Rendah: Skor terendah tampaknya sekitar 77, diamati pada 20/11/2022, 27/11/2022, dan 4/12/2022.

     Fluktuasi: Skor umumnya berfluktuasi antara poin tinggi dan rendah, yang menunjukkan variasi keterlibatan atau popularitas dari waktu ke waktu.

     Data Parsial: Ada kolom "isPartial" yang menunjukkan apakah data sebagian atau tidak. Entri terakhir pada 7/9/2023 ditandai sebagai "BENAR", menunjukkan bahwa data mungkin tidak lengkap untuk tanggal tersebut.
